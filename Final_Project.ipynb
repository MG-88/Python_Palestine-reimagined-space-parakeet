```python
import pandas as pd
import matplotlib.pyplot as plt

file_path  = '/workspaces/codespaces-blank/Palestinian Github Table - Sheet1.csv'
df = pd.read_csv(file_path)

fig, ax= plt.subplots(figsize=(9,5))
ax.axis('off')

table_data = df[['Killed', 'Wounded', 'Women/Children Killed', 'Housing Units Destroyed', 'Housing Units Damaged', 'Hospitals Destroyed' ]]
table = ax.table(cellText=table_data.values,
                 colLabels=table_data.columns,
                 cellLoc='center',
                 loc='bottom')

table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1.2, 1.2)

plt.suptitle('Palestine War Impact', y=0.16, fontsize=17, ha='center')

plt.show()
```
